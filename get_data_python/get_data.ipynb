```python
for k in range(1, 17):
    if(k == 4): 
        continue
    #k = 1
    cnt = str(k)
    url = cse+ae+cnt+ '?rcn=4&cra=%s&crb=%s' %(cra, crb)
    r = requests.get(url, headers=headers, data=payload)

    if r.status_code == 200:
        cin = r.json()['m2m:rsp']['m2m:cin']

        idx = 0
        for instance in cin:
            con = pd.Series(instance['con'].replace('\r', '').split(','))
            time_str = con[3] + ' ' + con[4]

            con_time = time.strptime(time_str, '%Y/%m/%d %H:%M:%S')
            con[13] = time.mktime(con_time)

            df.loc[idx] = con
            idx += 1
        df = df.sort_values(by=13)

    all_df = pd.concat([all_df, df])
```
